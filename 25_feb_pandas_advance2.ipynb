```python
import pandas as pd
import numpy as np

# Define the initial DataFrame
course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Q1: Write a code to print the data present in the second row of the DataFrame df.
print("Q1:")
print(df.iloc[1])

# Q2: What is the difference between the functions loc and iloc in pandas.DataFrame?
print("\nQ2:")
# loc: Label-based indexing; includes the end label
# iloc: Integer-location based indexing; excludes the end index

# Q3: Reindex the DataFrame using a variable, reindex = [3,0,1,2] and find output for new_df.loc[2] and new_df.iloc[2].
print("\nQ3:")
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)
print("new_df.loc[2]:")
print(new_df.loc[2])
print("new_df.iloc[2]:")
print(new_df.iloc[2])

# Q4: Statistical measurements for DataFrame df1
print("\nQ4:")
df1 = pd.DataFrame({
    'column_1': np.random.rand(6),
    'column_2': np.random.rand(6),
    'column_3': np.random.rand(6),
    'column_4': np.random.rand(6),
    'column_5': np.random.rand(6),
    'column_6': np.random.rand(6)
}, index=[1, 2, 3, 4, 5, 6])

# (i) Mean of each column
print("Mean of each column:")
print(df1.mean())

# (ii) Standard deviation of column 'column_2'
print("Standard deviation of 'column_2':")
print(df1['column_2'].std())

# Q5: Replace data in second row of column 'column_2' and find the mean of column 'column_2'
print("\nQ5:")
df1.loc[2, 'column_2'] = 'string_value'
try:
    mean_column_2 = df1['column_2'].astype(float).mean()
except ValueError as e:
    print("Error:", e)

# Q6: What are window functions in pandas and list the types of window functions?
print("\nQ6:")
# Window functions are used to perform operations over a rolling window of data.
# Types:
# 1. Rolling window: rolling()
# 2. Expanding window: expanding()
# 3. Exponential weighted window: ewm()

# Q7: Write a code to print only the current month and year.
print("\nQ7:")
current_date = pd.Timestamp.now()
print("Current Month and Year:", current_date.strftime('%B %Y'))

# Q8: Python program to calculate the difference between two dates.
print("\nQ8:")
date1 = pd.to_datetime(input("Enter the first date (YYYY-MM-DD): "))
date2 = pd.to_datetime(input("Enter the second date (YYYY-MM-DD): "))
delta = date2 - date1
print(f"Difference: {delta.days} days, {delta.components.hours} hours, {delta.components.minutes} minutes")

# Q9: Program to convert a column to categorical data type.
print("\nQ9:")
file_path = input("Enter the file path: ")
column_name = input("Enter the column name: ")
category_order = input("Enter the category order (comma-separated): ").split(',')

df_cat = pd.read_csv(file_path)
df_cat[column_name] = pd.Categorical(df_cat[column_name], categories=category_order, ordered=True)
print(df_cat.sort_values(by=column_name))

# Q10: Program to visualize sales data using a stacked bar chart.
print("\nQ10:")
import matplotlib.pyplot as plt

file_path = input("Enter the file path: ")
df_sales = pd.read_csv(file_path)
df_sales_pivot = df_sales.pivot_table(index='Date', columns='Product', values='Sales', aggfunc='sum')
df_sales_pivot.plot(kind='bar', stacked=True)
plt.title('Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()

# Q11: Program to calculate mean, median, and mode of test scores.
print("\nQ11:")
file_path = input("Enter the file path: ")
df_student = pd.read_csv(file_path)
mean_score = df_student['Test Score'].mean()
median_score = df_student['Test Score'].median()
mode_score = df_student['Test Score'].mode().tolist()

result_df = pd.DataFrame({
    'Statistic': ['Mean', 'Median', 'Mode'],
    'Value': [mean_score, median_score, ', '.join(map(str, mode_score))]
})
print(result_df)
```
