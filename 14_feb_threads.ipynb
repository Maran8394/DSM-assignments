```python
# Q1: What is multithreading in Python? Why is it used? Name the module used to handle threads in Python.

# Multithreading is a technique in programming where multiple threads run concurrently within a single process.
# It allows for parallel execution of tasks, which can lead to more efficient use of CPU resources and faster program execution.
# It is often used to perform tasks that are I/O bound or require waiting, allowing other tasks to proceed concurrently.

# The module used to handle threads in Python is `threading`.

# Q2: Why threading module used? Write the use of the following functions
# - activeCount()
# - currentThread()
# - enumerate()

# The `threading` module provides a way to create and manage threads in Python.

# - `activeCount()`: Returns the number of Thread objects currently alive.
# - `currentThread()`: Returns the current Thread object, corresponding to the caller's thread of control.
# - `enumerate()`: Returns a list of all Thread objects currently alive.

import threading

print("Active threads:", threading.activeCount())
print("Current thread:", threading.currentThread())
print("All threads:", threading.enumerate())

# Q3: Explain the following functions
# - run()
# - start()
# - join()
# - isAlive()

# - `run()`: Method that represents the code to be executed by a thread. It should be overridden in a subclass of Thread.
# - `start()`: Starts the thread's activity by calling the `run()` method in a new thread of control.
# - `join()`: Blocks the calling thread until the thread whose `join()` method is called terminates.
# - `isAlive()`: Returns whether the thread is alive (i.e., has been started and has not yet terminated).

# Q4: Write a Python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes.

import threading

def print_squares():
    squares = [x**2 for x in range(1, 6)]
    print("Squares:", squares)

def print_cubes():
    cubes = [x**3 for x in range(1, 6)]
    print("Cubes:", cubes)

# Creating thread objects
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Starting threads
thread1.start()
thread2.start()

# Waiting for both threads to complete
thread1.join()
thread2.join()

# Q5: State advantages and disadvantages of multithreading.

# Advantages:
# 1. Improved performance on multi-core processors.
# 2. Better resource utilization and responsiveness in I/O bound programs.
# 3. Simpler program structure for tasks that are naturally concurrent.

# Disadvantages:
# 1. Increased complexity in program design and debugging.
# 2. Potential for race conditions and deadlocks.
# 3. Threads share memory space, which can lead to synchronization issues.

# Q6: Explain deadlocks and race conditions.

# Deadlocks:
# A deadlock is a situation in which two or more threads are blocked forever, each waiting for the other to release a resource.
# Example: Two threads may hold locks on two resources and each thread is waiting for the resource held by the other.

# Race Conditions:
# A race condition occurs when two or more threads access shared resources concurrently and the outcome depends on the order of execution.
# Example: Two threads updating the same variable without proper synchronization can lead to inconsistent or incorrect results.
```
