```python
# Q1: What is an Exception in Python? Write the difference between Exceptions and Syntax Errors.

# Exception: An exception is an event that occurs during the execution of a program that disrupts the normal flow of instructions.
# Exceptions are typically used to handle errors or unusual conditions that arise during runtime.

# Difference between Exceptions and Syntax Errors:
# - Exceptions:
#   - Occur during the execution of the program.
#   - Examples include dividing by zero, accessing a file that doesn't exist, or trying to use an undefined variable.
#   - Handled using try and except blocks.

# - Syntax Errors:
#   - Occur when there is a mistake in the syntax of the code.
#   - Detected by the Python interpreter before the program runs.
#   - Examples include missing colons, mismatched parentheses, or incorrect indentation.
#   - Must be fixed before the code can be executed.

# Q2: What Happens When an Exception is Not Handled? Explain with an Example.

def divide(a, b):
    return a / b

try:
    print(divide(10, 0))  # This will cause a ZeroDivisionError
except ZeroDivisionError as e:
    print("An error occurred:", e)

# Q3: Which Python Statements Are Used to Catch and Handle Exceptions? Explain with an Example.

try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("Division successful")
finally:
    print("Execution complete")

# Q4: Explain with an Example: try and else, finally, and raise

class CustomError(Exception):
    pass

def divide(a, b):
    try:
        if b == 0:
            raise CustomError("Custom error: Cannot divide by zero")
        return a / b
    except CustomError as e:
        print(e)
    else:
        print("Division successful")
    finally:
        print("Execution complete")

print(divide(10, 0))  # CustomError will be raised and handled
print(divide(10, 2))  # Normal division

# Q5: What Are Custom Exceptions in Python? Why Do We Need Custom Exceptions? Explain with an Example.

class AgeError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def check_age(age):
    if age < 0:
        raise AgeError("Age cannot be negative")
    print("Age is valid")

try:
    check_age(-5)
except AgeError as e:
    print(e)

# Q6: Create a Custom Exception Class. Use This Class to Handle an Exception. Answer Should Be Simple and in Simple English.

class MyCustomError(Exception):
    def __init__(self, message):
        self.message = message
        super().__init__(self.message)

def do_something(value):
    if value < 0:
        raise MyCustomError("Value cannot be negative")
    print("Value is:", value)

try:
    do_something(-10)
except MyCustomError as e:
    print("Caught an error:", e)
```
